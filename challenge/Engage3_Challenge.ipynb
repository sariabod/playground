```python
# Let us take a look at pricing variance along a upc
```

```python
dfmin = df_pivot.min(axis=1).tolist()
```

```python
dfmax = df_pivot.max(axis=1).tolist()
```

```python
# lets check % diff
```

```python
diff_list = []
```

```python
for n, x in zip(dfmin, dfmax):
    diff = ((x - n) / ((x + n) / 2)) * 100
    diff_list.append(diff)
```

```python
diff_list.sort(reverse=True)
```

```python
diff_list[0:20]
```

```python
# These % difference between the min/max price is somthing I would dig into further. Regions will affect price, but having this many differences approaching 200%
# tells me there might a be a data quality issue. Either UPC are not correct or the agent is not entering the data correctly.
```
